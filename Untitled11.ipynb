```python
!pip install python-constraint

from constraint import Problem

# Initialize the problem
problem = Problem()

# Define the time slots and events
time_slots = ['T1', 'T2', 'T3', 'T4']
events = ['E1', 'E2', 'E3', 'E4']
speakers = ['S1', 'S2', 'S3', 'S4']

# Add variables for events (each event gets one time slot)
for event in events:
    problem.addVariable(event, time_slots)

# Add variables for speakers (each speaker must be assigned at least one event)
for speaker in speakers:
    problem.addVariable(speaker, events)

# Add constraints:

# 1. Event E1 must be scheduled in T1
problem.addConstraint(lambda t: t == 'T1', ('E1',))

# 2. Speaker S2 prefers T2 (soft preference)
# We'll add this as a hard constraint to force S2 to speak at T2
problem.addConstraint(lambda t: t == 'T2', ('S2',))

# 3. Events E2 and E3 must not be scheduled at the same time slot
def no_overlap_e2_e3(e2, e3):
    return e2 != e3
problem.addConstraint(no_overlap_e2_e3, ('E2', 'E3'))

# 4. No speaker can be assigned to more than one event at the same time
def no_conflict(speaker_event1, speaker_event2, speaker_event3, speaker_event4):
    return speaker_event1 != speaker_event2 and speaker_event1 != speaker_event3 and speaker_event1 != speaker_event4 \
           and speaker_event2 != speaker_event3 and speaker_event2 != speaker_event4 \
           and speaker_event3 != speaker_event4
problem.addConstraint(no_conflict, ('S1', 'S2', 'S3', 'S4'))

# Get the solution
solution = problem.getSolutions()

# Output the solution
for sol in solution:
    print("Solution:")
    for key, value in sol.items():
        print(f"{key} -> {value}")
```

```python
!pip install networkx matplotlib python-constraint

import networkx as nx
import matplotlib.pyplot as plt
from constraint import Problem

# Initialize the problem
problem = Problem()

# Define the regions (states/territories of Australia)
regions = ['WA', 'NT', 'SA', 'Q', 'NSW', 'V', 'T']

# Define the colors available for the regions
colors = ['C1', 'C2', 'C3']

# Add the variables for each region (with domains C1, C2, C3)
for region in regions:
    problem.addVariable(region, colors)

# Define adjacency between regions (which regions are adjacent)
adjacency = {
    'WA': ['NT', 'SA'],
    'NT': ['WA', 'SA', 'Q', 'NSW'],
    'SA': ['WA', 'NT', 'Q', 'NSW', 'V'],
    'Q': ['NT', 'SA', 'NSW'],
    'NSW': ['Q', 'SA', 'V'],
    'V': ['SA', 'NSW'],
    'T': []  # Tasmania is isolated
}

# Add constraints: No two adjacent regions can have the same color
for region1 in adjacency:
    for region2 in adjacency[region1]:
        problem.addConstraint(lambda r1, r2: r1 != r2, (region1, region2))

# Get the solutions
solution = problem.getSolutions()

# Output the solution
if solution:
    print("Solution found:")
    for sol in solution:
        for region in regions:
            print(f"{region}: {sol[region]}")
else:
    print("No solution found.")

# Let's create a graph to visualize the regions and adjacency
G = nx.Graph()

# Add the nodes (regions)
G.add_nodes_from(regions)

# Add edges based on adjacency
for region1 in adjacency:
    for region2 in adjacency[region1]:
        G.add_edge(region1, region2)

# Visualize the graph with the color assignments
color_map = []
for region in regions:
    if solution:
        color = solution[0][region]  # Just taking the first solution
        if color == 'C1':
            color_map.append('red')
        elif color == 'C2':
            color_map.append('blue')
        elif color == 'C3':
            color_map.append('green')

# Draw the graph
plt.figure(figsize=(10, 8))
pos = nx.spring_layout(G)  # For positioning nodes in a visually appealing way
nx.draw(G, pos, with_labels=True, node_size=20000, node_color=color_map, font_size=15, font_weight='bold', edge_color='gray')
plt.title('Australia Map Coloring Problem', fontsize=16)
plt.show()
```
